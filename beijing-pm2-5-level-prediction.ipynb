```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Assuming 'df' is your cleaned dataset
df['date'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])

# Set the index to the date column
df.set_index('date', inplace=True)

# Create time-based features
df['month'] = df.index.month
df['day'] = df.index.day
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek

# Lag features
n_lags = 3  # Number of lags
for lag in range(1, n_lags + 1):
    df[f'lag_{lag}'] = df['pm2.5'].shift(lag)

# Rolling statistics
df['rolling_mean_3'] = df['pm2.5'].rolling(window=3).mean()  # 3-hour rolling mean
df['rolling_mean_24'] = df['pm2.5'].rolling(window=24).mean()  # 24-hour rolling mean

# Differencing to remove trend
df['pm2.5_diff'] = df['pm2.5'].diff()

# Drop any rows with missing values created by shift or rolling
df.dropna(inplace=True)

# Select features for modeling
features = ['pm2.5', 'month', 'day', 'hour', 'day_of_week', 'lag_1', 'lag_2', 'lag_3', 'rolling_mean_3', 'rolling_mean_24', 'pm2.5_diff']

# Scale features for LSTM/GRU input
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features, index=df.index)
```

```python
def create_sequences(data, target_col, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data.iloc[i:i + sequence_length].values)  # Use .iloc for positional indexing
        y.append(data.iloc[i + sequence_length, target_col])  # Use .iloc for the target column
    return np.array(X), np.array(y)


# Use last column ('pm2.5') as the target
sequence_length = 24  # You can use longer sequences (e.g., for monthly/yearly trends)
X, y = create_sequences(df_scaled, target_col=0, sequence_length=sequence_length)

# Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

LSTMs

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout

# Define the model
model = Sequential()

# LSTM/GRU layers
model.add(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))  # Regularization to prevent overfitting
model.add(LSTM(units=64, return_sequences=False))
# If using GRU, replace the above LSTM layers with:
# model.add(GRU(units=64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
# model.add(GRU(units=64, return_sequences=False))

# Output layer
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
model.summary()
```

Monthly and Yearly Trends

```python
# Evaluate on test data
loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')

# Forecast future PM2.5 values
y_pred = model.predict(X_test)

# Inverse transform to get original scale
y_pred_rescaled = scaler.inverse_transform(np.concatenate([y_pred, np.zeros((y_pred.shape[0], len(features)-1))], axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate([y_test.reshape(-1, 1), np.zeros((y_test.shape[0], len(features)-1))], axis=1))[:, 0]

# Compare predictions to true values
import matplotlib.pyplot as plt

plt.plot(df.index[-len(y_test):], y_test_rescaled, label='True Monthly PM2.5')
plt.plot(df.index[-len(y_pred):], y_pred_rescaled, label='Predicted Monthly PM2.5')
plt.legend()
plt.show()
```

```python
# Resample to monthly average
df['predicted_pm2.5'] = np.nan
df.loc[df.index[-len(y_pred_rescaled):], 'predicted_pm2.5'] = y_pred_rescaled
monthly_trend = df[['pm2.5', 'predicted_pm2.5']].resample('ME').mean()

# Plot the monthly trend
plt.plot(monthly_trend.index, monthly_trend['pm2.5'], label='True yearly Avg PM2.5')
plt.plot(monthly_trend.index, monthly_trend['predicted_pm2.5'], label='Predicted yearly Avg PM2.5')
plt.legend()
plt.show()
```

Daily and hourly trends

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Assuming 'df' is your cleaned dataset
df['date'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
df.set_index('date', inplace=True)

# Time-based features for daily, hourly trends
df['month'] = df.index.month
df['day'] = df.index.day
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek

# Adding daily and hourly average PM2.5 as features
df['daily_avg_pm2.5'] = df['pm2.5'].resample('D').transform('mean')
df['hourly_avg_pm2.5'] = df['pm2.5'].resample('h').transform('mean')

# Lag features for daily and hourly trends
n_lags = 3
for lag in range(1, n_lags + 1):
    df[f'lag_{lag}_hour'] = df['pm2.5'].shift(lag)
    df[f'lag_{lag}_day'] = df['pm2.5'].shift(lag * 24)  # 24-hour lags for daily trends

# Rolling statistics for daily and hourly trends
df['rolling_mean_hour_3'] = df['pm2.5'].rolling(window=3).mean()  # 3-hour rolling mean
df['rolling_mean_day_7'] = df['pm2.5'].rolling(window=24*7).mean()  # 7-day rolling mean

# Drop missing values caused by rolling/lags
df.dropna(inplace=True)

# Select features for modeling
features = ['pm2.5', 'month', 'day', 'hour', 'day_of_week', 'daily_avg_pm2.5', 
            'hourly_avg_pm2.5', 'lag_1_hour', 'lag_2_hour', 'lag_3_hour', 
            'lag_1_day', 'lag_2_day', 'lag_3_day', 'rolling_mean_hour_3', 
            'rolling_mean_day_7']

# Scale the data
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features, index=df.index)
```

```python
def create_sequences(data, target_col, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data.iloc[i:i + sequence_length].values)  # Use .iloc for correct indexing
        y.append(data.iloc[i + sequence_length, target_col])  # Use .iloc for target
    return np.array(X), np.array(y)

# Sequence length of 24 hours (or longer for monthly/yearly trends)
sequence_length = 24
X, y = create_sequences(df_scaled, target_col=0, sequence_length=sequence_length)

# Split into train and test sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout, GRU

# LSTM/GRU Model for daily and hourly forecasting
model = Sequential()

# LSTM/GRU layers
model.add(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))  # Prevent overfitting
model.add(LSTM(units=64, return_sequences=False))

# If using GRU, you can replace with GRU layers:
# model.add(GRU(units=64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
# model.add(GRU(units=64, return_sequences=False))

# Output layer
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate model performance
loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')

# Predict on test data
y_pred = model.predict(X_test)

# Inverse transform to get original PM2.5 values
y_pred_rescaled = scaler.inverse_transform(np.concatenate([y_pred, np.zeros((y_pred.shape[0], len(features)-1))], axis=1))[:, 0]
y_test_rescaled = scaler.inverse_transform(np.concatenate([y_test.reshape(-1, 1), np.zeros((y_test.shape[0], len(features)-1))], axis=1))[:, 0]

# Plot actual vs predicted PM2.5 for daily/hourly trend visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df.index[-len(y_test):], y_test_rescaled, label='True PM2.5')
plt.plot(df.index[-len(y_pred):], y_pred_rescaled, label='Predicted PM2.5')
plt.title('True vs Predicted PM2.5 (Hourly Trend)')
plt.legend()
plt.show()
```
